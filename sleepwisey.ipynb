```python
import pandas as pd
import io
import requests
from pyquery import PyQuery as pq
linkf='https://taxfoundation.org/2020-sales-taxes/'
f = requests.get(linkf)
ffd= pq(f.text)

dictA = {}

for j in ffd('#post-content > table >tbody >tr' ).items():
    state = j('td:nth-child(1)').text()
    combine_rate = j('td:nth-child(5)').text()
    dictA[state] = combine_rate

dictA['California'] = '8.66%'
dictA['Hawaii'] = '4.44%'
dictA['New Jersey'] = '6.6%'
dictA['New Mexico'] = '7.82%'
dictA['South Dakota'] = '6.4%'
dictA['Utah'] = '7.18%'
dictA['Virginia'] = '5.65%'
dictA['other'] = '0%'

    
    
```

```python
#user should start from here
statename = input('Which state are you in \n')
#find the list for state tax
#dictA[statename]
#how much tips do you pay

tip_rate = input("tip_rate (%):(no tip please press 0)\n")
if type(statename) == str:
    thistimetax = (float(dictA[statename].replace("%","")))*0.01
    print(f"State tax rate is {dictA[statename]}")
    try:
        float(tip_rate)
        print(f'Your tip rate is {tip_rate}%')
        tiprate = float(tip_rate) * 0.01 
    except:    
        print("not a number")
```

```python
#group establish
member = int(input('How many members do you have '))
DictM = {}
listM = []
for K in range(member):
    membername =input("Tell me username: ")
    #需避免同名
    listM.append(membername)
    DictM['%s' %membername] =0
    #lambda 改寫
#we cannot have the same name
```

```python
#share equally
def share_equally():
    print(DictM.keys())
    who_pay_atfirst = input('Who pay this bill?\n ')
    bill_amount = float(input('How much?\n '))
    DictM[who_pay_atfirst] += bill_amount
    for q in range(member):
        DictM[listM[q]] -= round(bill_amount/member,2)

    
    #需要新增均分模式
    #誰給誰錢要文字提示
```

```python
#pay each
def pay_each():
    print(DictM.keys())
    who_pay = input("Who pay this bill?\n ")
    membername2 = str(input("Who need to pay?\n"))
    amount = float(input('How much? (pre-tax,pre-tip)\n')\
    *(1+thistimetax) * (1+tiprate),2)
    DictM[membername2] -= amount
    DictM[who_pay] += amount

    #involve state tax
    #need to involve not tax item.
```

```python
#menu receipt
checkpoint = False
while True:
    print('(a)Share equally\n')
        # 除法了後要每個人分得一樣

    print("(b)Assign people to pay\n")
    print('(c)End & Result')
    order = input()
    if order == 'a':
        share_equally()
    elif order == 'b':
        pay_each()
    elif order == 'c':
        print(DictM.items())
        
        
        borrow,lend ={},{}
        temp_list = sorted(list(DictM.items()))
        for i in range(len(temp_list)):
            if temp_list[i][1] >=0 :
                lend[temp_list[i][0]] = temp_list[i][1]
            elif temp_list[i][1] < 0:
                borrow[temp_list[i][0]] = temp_list[i][1]
        temp_borrow = list(borrow.values())
        temp_lend = list(lend.values())
        
             
      
        for K in range(len(borrow)):
            #wont start new if borrow[k] is not zero
            
            while abs(temp_borrow[K]) > 0:
                # the most number will be len(borrow) *len(lend)
                for j in range(len(lend)):
                    if abs(temp_borrow[K]) >= temp_lend[j]:
                        #把位置先存起來
                        #會出現付0元的現象
                        loc_b = K
                        loc_l = j
                        print(f'%s Pay' %list(borrow.keys())[K],f'%s' %list(lend.keys())[j],\
                             "{:.2f}".format(abs(temp_lend[j])))
                        temp_borrow[K] += temp_lend[j]
                        temp_lend[j] = 0
                        
                    elif abs(temp_borrow[K]) < temp_lend[j]:
                        loc_b = K
                        loc_l = j
                        print(f'%s Pay' %list(borrow.keys())[K],f'%s' %list(lend.keys())[j], \
                              "{:.2f}".format(abs(temp_borrow[K])))

                        temp_lend[j] += temp_borrow[K]
                        temp_borrow[K] = 0
                if (K == len(borrow)-1 and j == len(lend))-1 or temp_borrow[len(temp_borrow)-1] ==0:
                    checkpoint = True
                    break
                        

        if checkpoint == True:
            break
#技術總結
#While選單
#if (c)end
#分borrow and lend
#while 迴圈直到borrow歸零
#
#for 借 檢查借
# for 欠 檢查欠, A*B 是最大可能
#宣判 誰付誰錢
#if 來做結束 （連結到temp_borrow 跟temp_lend的最後一個）
```
